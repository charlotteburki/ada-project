```python
def apply_genre(df, venue, genre) :
    
    for index in total_events[total_events['Venue']==venue].index :
        if  pd.isnull(total_events.loc[index].genre) :
            total_events.set_value(index, 'genre', genre)
        elif ((genre == 'Electro') & ((total_events.loc[index].genre == 'Rock') | (total_events.loc[index].genre == 'Classical')) ) :
            total_events.set_value(index, 'genre', genre)
    return
```
