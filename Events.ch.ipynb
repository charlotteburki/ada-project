```python
#Create df
columns = ['Artist', 'Date','Venue', 'Genre']
df = pd.DataFrame(columns=columns)
df
```

```python
#Parse future events
for i in range(1,265):
    print(i)
    baseFormUrl = "https://events.ch/de/Konzerte/" + str(i)
    r = requests.get(baseFormUrl)
    soup = BeautifulSoup(r.content, "html.parser")
   
    while True:
        try:
            event = soup.find('div', class_='event-main-information').extract()
        except Exception as inst:
                break
        date = event.meta['content']
        venue = event.find('li', itemprop="location" ).a.string
        genre = event.find('ul', class_="categories-list").li.next_element.next_element.next_element
        artist = event.a.next_element.next_element.string
        df = df.append(pd.DataFrame([[artist, date, venue, genre]],columns=columns) )
       
```
